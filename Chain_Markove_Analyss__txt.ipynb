## import libraries

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
import pandas as pd
import numpy as np
import random
from collections import defaultdict
```

```python
# Fetch HTML content from the URL
html = urlopen("https://en.wikipedia.org/wiki/Rabat").read()
```

## Scraping the text

```python
# Parse HTML and extract text
soup = BeautifulSoup(html, "html.parser")
text = ''.join(soup.stripped_strings)
text
```

```python
lowercase_text = text.lower()
# Replace multiple spaces with a single space
clean_text = re.sub(r'\s+', ' ', lowercase_text)
# Remove all characters except letters and spaces
clean_text = re.sub(r'[^a-z\s]','', lowercase_text)
```

```python

# Initialize a dictionary to store letter counts
letter_count = {}
```

```python

for char in clean_text:
    if char.isalpha() or char.isspace():
        if char in letter_count:
            letter_count[char] += 1
        else:
            letter_count[char] = 1


for letter, count in letter_count.items():
    print(f"{letter}: {count}")
```

```python

letter_counts = {}

space_count1 = clean_text.count(' ')

for char in clean_text:
    if char.isalpha():
        if char in letter_counts:
            letter_counts[char] += 1
        else:
            letter_counts[char] = 1

for letter, count in letter_counts.items():
    print(f"{letter}: {count}")

print(f"Space: {space_count1}")
L = [count for letter , count in letter_counts.items()]
np.sum(L)+np.sum(space_count1)
```

## calculate the matrix of transition

```python

texte = clean_text  # Utiliser le texte nettoyé précédemment

for i in range(len(texte) - 1):

    x = ord(texte[i]) - ord('a') if texte[i].isalpha() else 26
    y = ord(texte[i + 1]) - ord('a') if texte[i + 1].isalpha() else 26
    M[x,y] += 1

# Afficher la matrice M
#for row in M:
    #print(row/(np.sum(row)))
P=np.zeros((27,27))
M[0,:]
for i in range(27):
  P[i,:]=M[i,:]/(np.sum(M[i,:]))
```

```python

score = 0
attempts = 0

# Second text for testing (replace this with your actual second text)
second_text = "The quick brown fox jumps over the lazy dog."



for i in range(len(second_text) - 1):
    x = ord(second_text[i]) - ord('a') if second_text[i].isalpha() else 26
    y = ord(second_text[i + 1]) - ord('a') if second_text[i + 1].isalpha() else 26

    transition_prob = P[x, y]

    score += transition_prob

    attempts += 1

overall_accuracy = score / attempts

# Print the overall accuracy
print("Overall Accuracy:", overall_accuracy)
```

```python
score = 0
attempts = 0

for i in range(len(texte)-3):
    current_state = texte[i:i+3]
    next_state = texte[i+3]
    if next_state in transition_matrix[current_state]:
        score += 1
    attempts += 1

accuracy = score / attempts
print("Performance du modèle de Markov d'ordre 3:", accuracy)
```

###  Modélisation du texte comme succession de mots

```python

texte = clean_text

mots = texte.split()
mots = [''.join(filter(str.isalpha, mot)) for mot in mots]

paires_de_mots = [(mots[i], mots[i+1]) for i in range(len(mots)-1)]


transition_matrix_mots = defaultdict(lambda: defaultdict(int))

for i in range(len(mots)-1):
    transition_matrix_mots[mots[i]][mots[i+1]] += 1

for mot, transitions in transition_matrix_mots.items():
    total_transitions = sum(transitions.values())
    for mot_suivant in transitions:
        transition_matrix_mots[mot][mot_suivant] /= total_transitions



```

```python
score = 0
attempts = 0

for i in range(len(mots)-1):
    mot_actuel = mots[i]
    mot_suivant = mots[i+1]
    if mot_suivant in transition_matrix_mots[mot_actuel]:
        score += 1
    attempts += 1

accuracy = score / attempts
print("Performance du modèle de succession de mots:", accuracy)
```

###  Génération de texte avec un modèle de Markov:

```python


etat_initial = "rabat"

nombre_mots = 50

texte_genere = [etat_initial]

for _ in range(nombre_mots):

    dernier_mot = texte_genere[-1]

    mots_suivants = list(transition_matrix_mots[dernier_mot].keys())
    probabilites = list(transition_matrix_mots[dernier_mot].values())
    mot_suivant = random.choices(mots_suivants, weights=probabilites)[0]


    texte_genere.append(mot_suivant)

#
texte_genere = ' '.join(texte_genere)
print(texte_genere)
```
